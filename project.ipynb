## Data Cleaning

```python
data['Embarked']=data['Embarked'].replace({'S':'Southampton','C':'Chebourg','Q':'Queenstown'})
```

```python
data.head()
```

```python
data['Embarked'].value_counts()
```

```python
data.dropna(subset=['Embarked'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data['Fare'] = data['Fare'].astype('int8')
```

```python
data.head()
```

## Feature Encoding

```python
## Encoder -- Gender ---> LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
label = LabelEncoder()
```

```python
data['Sex']=label.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse_output=False)
```

```python
data.columns
```

```python
Embarked = onehot.fit_transform(data[['Embarked']])
```

```python
Embarked = pd.DataFrame(Embarked, columns=onehot.get_feature_names_out())
```

```python
Embarked
```

```python
data = pd.concat([data.drop(columns=['Embarked']), Embarked],axis=True)
```

```python
data.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler() 
```
