## Train-Validation-Test split

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data= data.dropna()
```

```python
X=data.drop(columns='Survived')
```

```python
y=data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=20, random_state = 42)
```

```python
X_train, X_valid, y_train, y_valid=train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
import tensorflow
```

 ## Model Building

```python
(X_train.shape[1],)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential([Dense(128, input_shape=(X_train.shape[1],), activation='relu'),# First hidden Layer
Dense(64, activation='relu'),
Dense(32, activation='relu'),
Dense(1,activation='sigmoid')])
```

```python
model.summary()
```

## Model Compilaion ---> Optimizer, Loss, Metrics

```python
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
stopping_callback = EarlyStopping(monitor='val_loss', patience=5,restore_best_weights=True)
```

```python
# X_train, X_valid, y_train, y_valid
```

```python
model.fit(X_train,y_train,validation_data=(X_valid, y_valid), epochs=10, callbacks=[stopping_callback])
```

```python
model.save('model.h5')
```
